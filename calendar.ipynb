```python
import pandas as pd
from datetime import date, timedelta
```

```python
cols = [
    "gregorian_date",
    "gregorian_year",
    "gregorian_month",	
    "gregorian_day",	
    "retail_454_dayofweek",	
    "retail_454_dayofmonth",	
    "retail_454_month",	
    "retail_454_weekofmonth",	
    "retail_454_weeksinmonth",	
    "retail_454_weekofyear",	
    "retail_454_quarter",	
    "retail_454_year"
]
```

```python
gregorian_53_week_years = [2001, 2007, 2013, 2028, 2024]
nrf_53_week_years = [2000, 2006, 2012, 2017, 2023]
```

```python
df = pd.DataFrame(columns=cols)
```

```python
start_date = date(2000,1,30)
# end_date = date(2025,1,30)
# start_date = date(2001,1,7)
# end_date = date(2001,1,10)
end_date = date(2025,1,30)
df['gregorian_date'] = pd.date_range(start_date,end_date-timedelta(days=1),freq='d')
df['gregorian_year'] = pd.DatetimeIndex(df['gregorian_date']).year
df['gregorian_month'] = pd.DatetimeIndex(df['gregorian_date']).month
df['gregorian_day'] = pd.DatetimeIndex(df['gregorian_date']).day
df
```

```python
for i in range(len(df)):
    dt = df.loc[i, 'gregorian_date']
    year = dt.year
    cur_month = dt.month
    cur_date = dt.day

    """
        January is screwed up because in the NRF calendar it's the 12th month, but exactly when the 11th month ends and the 12th month begins varies.
        So, if the gregorian date is January, need to figure out what the "last day" of the previous year is so I can get the number of days in the "year"
        and use that to determine when the 12th month starts
    """
    if cur_month == 1:
        prev_jan_31 = date(year-1, 1, 31)
        prev_jan_31_wd = prev_jan_31.weekday()
        prev_jan_31_dow = prev_jan_31_wd + 2 if prev_jan_31_wd < 6 else prev_jan_31_wd - 5# Adding 2 because 0 is monday and I want Sunday to be 1 and Saturday to be 7
        if prev_jan_31_dow in [1,2,3]:
            fdopy = date(year - 1, 1, 31 - prev_jan_31_dow + 1)
        else:
            fdopy = date(year - 1, 2, 7 - prev_jan_31_dow + 1)

    # Get the "First" day of the year by
    # 1. Determine the closest Saturday to Jan 31st.
    jan_31 = date(year, 1, 31)
    jan_31_wd = jan_31.weekday()
    jan_31_dow = jan_31_wd + 2 if jan_31_wd < 6 else jan_31_wd - 5# Adding 2 because 0 is monday and I want Sunday to be 1 and Saturday to be 7
    if jan_31_dow in [1,2,3]:
        fdoy = date(year, 1, 31 - jan_31_dow + 1)
    else:
        fdoy = date(year, 2, 7 - jan_31_dow + 1)

    ldopy = fdoy - timedelta(days=1)

    cur_dow = dt.weekday()
    dow = cur_dow + 2 if cur_dow < 6 else cur_dow - 5
    doy = (dt.date() - fdoy).days + 1
    days_in_year = 52 * 7 if year in gregorian_53_week_years else 53 * 7
    ldoy = fdoy + timedelta(days=days_in_year)

    # Generate all the First days of the "month"
    fdom2 = fdoy + timedelta(days=7*4)
    fdom3 = fdom2 + timedelta(days=7*5)
    fdom4 = fdom3 + timedelta(days=7*4)
    fdom5 = fdom4 + timedelta(days=7*4)
    fdom6 = fdom5 + timedelta(days=7*5)
    fdom7 = fdom6 + timedelta(days=7*4)
    fdom8 = fdom7 + timedelta(days=7*4)
    fdom9 = fdom8 + timedelta(days=7*5)
    fdom10 = fdom9 + timedelta(days=7*4)
    fdom11 = fdom10 + timedelta(days=7*4)
    fdom12 = fdom11 + timedelta(days=7*5)

    prev_year = df.loc[i-1, 'retail_454_year'] if i > 0 else year
    prev_dom = df.loc[i-1, 'retail_454_dayofmonth'] if i > 0 else 1
    prev_month = df.loc[i-1, 'retail_454_month'] if i > 0 else 1
     
    # The day of the week is independant of any other dates so go ahead and set it. 
    df.at[i, 'retail_454_dayofweek'] = dow     
    if dt <= ldopy:
        dom = prev_dom + 1 if prev_dom < 35 else 1
        # Since there are 5 weeks in the 11th month there are 35 days
        month = 11 if prev_dom < (7 * 5) and prev_month == 11 else 12
        doy = (dt.date() - fdopy).days + 1      
        
        # The 11th month has 5 weeks and the 12th has 4, unless it's a "restated" year, then the 12th month gets 5 weeks
        weeks_in_month = 4
        if month == 11:
            weeks_in_month = 5
        else:
            if year in gregorian_53_week_years:
                weeks_in_month = 5

        df.at[i, 'retail_454_dayofmonth'] = dom
        df.at[i, 'retail_454_month'] = month 
        df.at[i, 'retail_454_weekofmonth'] = (dom + 6) // 7
        df.at[i, 'retail_454_weeksinmonth'] = weeks_in_month   
        df.at[i, 'retail_454_weekofyear'] = ((doy + 6) // 7)
        df.at[i, 'retail_454_quarter'] = 4
        df.at[i, 'retail_454_year'] = prev_year       
    elif dt == fdoy:
        df.at[i, 'retail_454_dayofmonth'] = 1
        df.at[i, 'retail_454_month'] = 1
        df.at[i, 'retail_454_weekofmonth'] = 1
        df.at[i, 'retail_454_weeksinmonth'] = 4
        df.at[i, 'retail_454_weekofyear'] = 1
        df.at[i, 'retail_454_quarter'] = 1
        df.at[i, 'retail_454_year'] = year        
    # 1st Month
    elif dt < fdom2:
        df.at[i, 'retail_454_dayofmonth'] = doy
        df.at[i, 'retail_454_month'] = 1
        df.at[i, 'retail_454_weekofmonth'] = (doy + 6) // 7
        df.at[i, 'retail_454_weeksinmonth'] = 4
        df.at[i, 'retail_454_weekofyear'] = (doy + 6) // 7
        df.at[i, 'retail_454_quarter'] = 1
        df.at[i, 'retail_454_year'] = prev_year   
    # 2nd Month
    elif dt < fdom3:
        dom = prev_dom + 1 if prev_month == 2 else 1
        df.at[i, 'retail_454_dayofmonth'] = dom
        df.at[i, 'retail_454_month'] = 2
        df.at[i, 'retail_454_weekofmonth'] = (dom + 6) // 7
        df.at[i, 'retail_454_weeksinmonth'] = 5 
        df.at[i, 'retail_454_weekofyear'] = ((doy + 6) // 7)
        df.at[i, 'retail_454_quarter'] = 1
        df.at[i, 'retail_454_year'] = prev_year         
    # 3rd Month
    elif dt < fdom4:
        dom = prev_dom + 1 if prev_month == 3 else 1
        df.at[i, 'retail_454_dayofmonth'] = dom
        df.at[i, 'retail_454_month'] = 3
        df.at[i, 'retail_454_weekofmonth'] = (dom + 6) // 7
        df.at[i, 'retail_454_weeksinmonth'] = 4    
        df.at[i, 'retail_454_weekofyear'] = ((doy + 6) // 7)
        df.at[i, 'retail_454_quarter'] = 1
        df.at[i, 'retail_454_year'] = prev_year   
    # 4th Month
    elif dt < fdom5: 
        dom = prev_dom + 1 if prev_month == 4 else 1
        df.at[i, 'retail_454_dayofmonth'] = dom
        df.at[i, 'retail_454_month'] = 4
        df.at[i, 'retail_454_weekofmonth'] = (dom + 6) // 7
        df.at[i, 'retail_454_weeksinmonth'] = 4    
        df.at[i, 'retail_454_weekofyear'] = ((doy + 6) // 7)
        df.at[i, 'retail_454_quarter'] = 2
        df.at[i, 'retail_454_year'] = prev_year    
    # 5th Month 
    elif dt < fdom6: 
        dom = prev_dom + 1 if prev_month == 5 else 1
        df.at[i, 'retail_454_dayofmonth'] = dom
        df.at[i, 'retail_454_month'] = 5
        df.at[i, 'retail_454_weekofmonth'] = (dom + 6) // 7
        df.at[i, 'retail_454_weeksinmonth'] = 5    
        df.at[i, 'retail_454_weekofyear'] = ((doy + 6) // 7)
        df.at[i, 'retail_454_quarter'] = 2
        df.at[i, 'retail_454_year'] = prev_year 
    # 6th Month      
    elif dt < fdom7:
        dom = prev_dom + 1 if prev_month == 6 else 1
        df.at[i, 'retail_454_dayofmonth'] = dom
        df.at[i, 'retail_454_month'] = 6
        df.at[i, 'retail_454_weekofmonth'] = (dom + 6) // 7
        df.at[i, 'retail_454_weeksinmonth'] = 4    
        df.at[i, 'retail_454_weekofyear'] = ((doy + 6) // 7)
        df.at[i, 'retail_454_quarter'] = 2
        df.at[i, 'retail_454_year'] = prev_year 
    # 7th Month   
    elif dt < fdom8: 
        dom = prev_dom + 1 if prev_month == 7 else 1     
        df.at[i, 'retail_454_dayofmonth'] = dom
        df.at[i, 'retail_454_month'] = 7
        df.at[i, 'retail_454_weekofmonth'] = (dom + 6) // 7
        df.at[i, 'retail_454_weeksinmonth'] = 4    
        df.at[i, 'retail_454_weekofyear'] = ((doy + 6) // 7)
        df.at[i, 'retail_454_quarter'] = 3
        df.at[i, 'retail_454_year'] = prev_year 
    # 8th Month     
    elif dt < fdom9: 
        dom = prev_dom + 1 if prev_month == 8 else 1        
        df.at[i, 'retail_454_dayofmonth'] = dom
        df.at[i, 'retail_454_month'] = 8
        df.at[i, 'retail_454_weekofmonth'] = (dom + 6) // 7
        df.at[i, 'retail_454_weeksinmonth'] = 5    
        df.at[i, 'retail_454_weekofyear'] = ((doy + 6) // 7)
        df.at[i, 'retail_454_quarter'] = 3
        df.at[i, 'retail_454_year'] = prev_year 
    # 9th Month 
    elif dt < fdom10: 
        dom = prev_dom + 1 if prev_month == 9 else 1         
        df.at[i, 'retail_454_dayofmonth'] = dom
        df.at[i, 'retail_454_month'] = 9
        df.at[i, 'retail_454_weekofmonth'] = (dom + 6) // 7
        df.at[i, 'retail_454_weeksinmonth'] = 4    
        df.at[i, 'retail_454_weekofyear'] = ((doy + 6) // 7)
        df.at[i, 'retail_454_quarter'] = 3
        df.at[i, 'retail_454_year'] = prev_year 
    # 10th Month 
    elif dt < fdom11: 
        dom = prev_dom + 1 if prev_month == 10 else 1        
        df.at[i, 'retail_454_dayofmonth'] = dom
        df.at[i, 'retail_454_month'] = 10
        df.at[i, 'retail_454_weekofmonth'] = (dom + 6) // 7
        df.at[i, 'retail_454_weeksinmonth'] = 4    
        df.at[i, 'retail_454_weekofyear'] = ((doy + 6) // 7)
        df.at[i, 'retail_454_quarter'] = 4
        df.at[i, 'retail_454_year'] = prev_year 
    # 11th Month 
    elif dt < fdom12: 
        dom = prev_dom + 1 if prev_month == 11 else 1         
        df.at[i, 'retail_454_dayofmonth'] = dom
        df.at[i, 'retail_454_month'] = 11
        df.at[i, 'retail_454_weekofmonth'] = (dom + 6) // 7
        df.at[i, 'retail_454_weeksinmonth'] = 5    
        df.at[i, 'retail_454_weekofyear'] = ((doy + 6) // 7)
        df.at[i, 'retail_454_quarter'] = 4
        df.at[i, 'retail_454_year'] = prev_year   
    # 12th Month 
    elif dt < ldoy: 
        print('In Else: ', dt, ' ', ldoy)
        dom = prev_dom + 1 if prev_month == 12 else 1        
        df.at[i, 'retail_454_dayofmonth'] = dom
        df.at[i, 'retail_454_month'] = 12
        df.at[i, 'retail_454_weekofmonth'] = (dom + 6) // 7
        df.at[i, 'retail_454_weeksinmonth'] = 5    
        df.at[i, 'retail_454_weekofyear'] = ((doy + 6) // 7)
        df.at[i, 'retail_454_quarter'] = 4
        df.at[i, 'retail_454_year'] = prev_year
    else:
        print('Need to figure out what to do if we get here')                                            
```
